# EDA USING BIVARIATE ANALYSIS

```python
import pandas as pd
import seaborn as sns
```

```python
tips= sns.load_dataset("tips")
titanic = pd.read_csv("train.csv")
flights = sns.load_dataset('flights')
iris = sns.load_dataset("iris")
```

### numerical -numerical

```python
tips
```

```python
sns.relplot(data=tips, x="total_bill",y="tip",hue="sex",style="smoker",size="size",kind="scatter")
```

# numerical - categorical

```python
titanic.sample(10)
```

```python
sns.catplot(kind="bar",data=titanic,x="Pclass",y="Age",hue="Sex")
```

```python
sns.boxplot(data=titanic,x="Sex",y="Age",hue="Survived")
```

```python
sns.displot(data=titanic,x="Age",kind="kde",hue="Survived")

# sns.distplot(titanic["Age"])
```

# categorical - categorical

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
titanic.groupby(["Embarked","Pclass"])["Survived"].sum()
```
