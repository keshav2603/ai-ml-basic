# One Hot Encoding

```python
import numpy as np 
import pandas as pd

```

```python
df= pd.read_csv("cars.csv")
df.sample(10)
```

```python
df["brand"].value_counts()
```

## one hot encoding using pandas

```python
pd.get_dummies(df,columns=["owner","fuel"],dtype=int)
```

## k-1 encoding


```python
pd.get_dummies(df,columns=["owner","fuel"],dtype=int,drop_first=True)
```

# using sklean

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
x_train, x_test, y_train,y_test = train_test_split(df.iloc[:,:4],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
x_train
```

```python
ohe=OneHotEncoder(drop="first",sparse_output=False)
```

```python
ohe.fit(x_train[["fuel","owner"]])
```

```python
x_train_encoded= ohe.transform(x_train[["fuel","owner"]])
x_test_encoded= ohe.transform(x_test[["fuel","owner"]])

x_train_encoded
```
